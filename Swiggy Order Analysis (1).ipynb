```python
country_wise_grouped = df.groupby('Country') # Grouping the restaurant with along with countries can give us better insights
country_wise_grouped.get_group('India')
```

```python
print(len(country_wise_grouped.get_group("India"))/len(df))
```

```python
Swiggy_Operational_Country = df.Country.unique()
print(Swiggy_Operational_Country)
print(country_wise_grouped)
```

```python
exposure=[]
for i in Swiggy_Operational_Country:
    ob = (len(country_wise_grouped.get_group(i))/len(df))*100
    exposure.append(ob)
Country_wise_exposure = pd.DataFrame({"Country Name":Swiggy_Operational_Country,"% Exposure": exposure})
Country_wise_exposure
```

```python
top_3 = Country_wise_exposure.nlargest(3,"% Exposure")
```

```python
plt.figure(figsize=(8,4))
ax = plt.pie(top_3["% Exposure"], labels=top_3["Country Name"],autopct='%1.1f%%')
plt.title("Top 3 Countries",fontweight="bold")
plt.show()
```

Here we can see that most of the exposure of Swiggy are there in India, United States and United Kingdom.

Popular Cuisines in India

```python
df["Cuisines"]=df["Cuisines"].str.split(",")
```

```python
## Now we will find out list of all the cuisines available in india.
```

```python
indian_cuisines = []
for i in df["Cuisines"]:
    for j in i:
        if j.strip() not in indian_cuisines:
            indian_cuisines.append(j.strip())
indian_cuisines
```

```python
len(indian_cuisines)
```

# Prediction of Votes

```python
data = df[df['Country']=='India']
data.head(1)
```

```python
data.head(1)
```

```python
data.drop(['Restaurant ID','Restaurant Name','Country','City','Address','Locality','Locality Verbose','Cuisines','Currency','Switch to order menu','Rating color'],axis=1,inplace=True)
```

```python
data = pd.get_dummies(data,drop_first=True)
data
```

```python
data.head(1)
```

```python
# Separate our dependent data and independent data
x = data.iloc[:,[0,1,2,3,5,6,7,8,9,10,11,12,13]].values
y = data.iloc[:,4].values
```

```python
x
```

```python
y
```

Train Test Split

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
y_train
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
y_test
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
